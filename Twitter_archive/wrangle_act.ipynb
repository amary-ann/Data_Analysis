### Issue #9:

#### Define
- Convert the rating numerator column to float and extract missing values from the text column

#### Code

```python
#Convert the column to float
df_twitter_archive['rating_numerator'] = df_twitter_archive['rating_numerator'].astype(float)
#Reading the float numerator values from text column  to a temporary column
df_twitter_archive['rating_num_temp'] = df_twitter_archive[df_twitter_archive.text.str.contains(r"(\d+\.\d\/\d+)")].text.str.extract(r"(\d+\.\d+)")
#Replacing the values to the original values
df_twitter_archive.loc[df_twitter_archive['rating_num_temp'].notnull(),['rating_numerator']] = df_twitter_archive['rating_num_temp']
#Dropping the temporary column
df_twitter_archive.drop(['rating_num_temp'], axis=1, inplace=True)
```

### Issue #11:

#### Define
- Merge the dog stages into one column

#### Code

```python
#Replacing the None values with blanks
df_twitter_archive.doggo.replace('None', '', inplace=True)
df_twitter_archive.floofer.replace('None', '', inplace=True)
df_twitter_archive.pupper.replace('None', '', inplace=True)
df_twitter_archive.puppo.replace('None', '', inplace=True)
```

```python
#Combining the four columns into one stage column
df_twitter_archive['stage'] = df_twitter_archive.doggo + df_twitter_archive.floofer + df_twitter_archive.pupper + df_twitter_archive.pupper + df_twitter_archive.puppo
```

```python
#Multiple stage values are assigned standard values for each multiple
df_twitter_archive.loc[df_twitter_archive.stage == 'doggopupper', 'stage'] = 'doggo,pupper'
df_twitter_archive.loc[df_twitter_archive.stage == 'doggopuppo', 'stage'] = 'doggo,puppo'
df_twitter_archive.loc[df_twitter_archive.stage =='doggofloofer', 'stage'] = 'doggo,floofer'
```

```python
#Dropping the columns
df_twitter_archive.drop(['doggo', 'floofer', 'pupper','puppo'], axis = 1, inplace = True)
```
