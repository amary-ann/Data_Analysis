## Investigation Overview

> The goal of this presentation is to expantiate more on the users of this service mainly using the user type feature. It is the find out the typical characteristics of those who make frequent and longer use in terms of duration of this service.


## Dataset Overview

> Ford GoBike System Data is a data set containing information about the individual rides made in a bike-sharing system covering the greater San Francisco Bay area.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('ford_go_bike.csv')
```

```python
#Covert datatype using datetime()
#start_time
df['start_time'] = pd.to_datetime(df['start_time'])

#end_time
df['end_time'] = pd.to_datetime(df['end_time'])
```

```python
#Derive the hour, day, month and year from both columns and store them in new columns
#start_time
df['start_hour'] = df['start_time'].dt.hour
df['start_day'] = df['start_time'].dt.day_name()


#end_time
df['end_hour'] = df['end_time'].dt.hour
df['end_day'] = df['end_time'].dt.day_name()
```

## Users of this service

>Plotting the percentage of users who use this service. There are more subsribers than customers who use this service


```python
#Plotting a pie chart of the user types
usr_type = df.user_type.value_counts()
plt.figure(figsize = (8,8))
plt.pie(usr_type, labels = usr_type.index,autopct='%0.0f%%', colors = ['skyblue', 'pink'])
plt.title("Subscribers vs Customers")
plt.show();
```

## Users with longer use of this service

> Plotting the duration of the user's trips. Customers take longer trips than subscribers in genera

```python
#Plot the duration of trips by user type 
ax = plt.figure(figsize = (20,10))
ax = sb.boxplot(x = 'user_type' , y = 'duration_sec', data = df, palette = 'coolwarm');
ax.set_ylim(1, 2500)
ax = plt.xlabel('User Type ', fontsize = 10)
ax = plt.ylabel('Duration in seconds', fontsize = 10)
ax = sb.despine()
ax = plt.title('Trip Duration of user type', fontsize = 20);
```

## Users who frequently use this service each day 
> Plotting the frequency of trips for the users for the days of the week

```python
#### Plot the frequency of user types for each day of the week
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
cx = plt.figure(figsize = (20,10))
cx = sb.countplot(x = 'start_day', hue = 'user_type', data = df, order = days, palette = 'coolwarm');
cx = sb.despine()
cx = plt.xticks(rotation = 45)
cx = plt.title('Trip Frequency of Users by Day ', fontsize = 20);
cx = plt.xlabel('Days of the Week', fontsize = 10)
cx = plt.ylabel('Count', fontsize = 10)
```

## Duration of use each day by users

>Plotting the duration of trips for the users for the days of the week. Customers take longer trips than subscribers

```python
#Plotting the duration of trips by users for each day of the week
dx = plt.figure(figsize = (20,10))
dx = sb.barplot(x = 'start_day',y = 'duration_sec', hue = 'user_type', data = df, palette= 'coolwarm');
dx.set_ylim(1,2500)
dx = sb.despine()
dx = plt.title('Duration of user trips for each day of the week', fontsize = 20)
```
